### Covariance Matrix

```python
covariance_matrix=crime_data.cov()
covariance_matrix.values
np.shape(covariance_matrix)
```

### Eigen Values: 

The top 20 Eigen values have been displayed in the table below. 
The 1st and 2nd eigen values seem to capture most of the information from the given data set and hence the first two eigen vectors can be taken as the principal components.

```python
eig_vals, eig_vecs = np.linalg.eig(covariance_matrix.values)
eig_vals=np.sort(eig_vals)
eig_vals=eig_vals[::-1]
top_20_values=pd.DataFrame(eig_vals[0:20],columns=['Eigen Values'])
top_20_values.index+=1
top_20_values
```
